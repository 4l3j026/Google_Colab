Un docente quiere realizar una actividad en grupo para una clase de n estudiantes, con exactamente x estudiantes por grupo. Sabe que es posible que la división implique la existencia de un grupo con menos estudiantes, por lo que plantea una estrategia.


En primer lugar, cuando vaya a realizar la actividad, asignará aleatoriamente los grupos para prevenir que sean los mismos estudiantes en cada actividad. En caso de que uno de los grupos quede con menos de x estudiantes, estos no realizarán la actividad y en su lugar apoyarán el proceso de calificación del resto de grupos.


Este grupo de calificadores deberá preparar y entregar un reporte corto de observaciones del trabajo realizado por cada uno de los grupos completos que fueron asignados.


Por ejemplo, para una clase con 16 estudiantes y una actividad que requiera crear grupos de 6 personas, podríamos obtener 2 grupos completos y un grupo de 4 calificadores. Cada uno de ellos calificaría individualmente a los 2 grupos creados, para un total de 8 reportes creados y entregados al docente.

Antes de pensar en cómo realizar la asignación, el docente quiere resolver las siguientes preguntas:


1. ¿Cuántos grupos completos de x estudiantes pueden armarse en una clase de n  estudiantes?
2. ¿Cuántos reportes en total debería recibir del grupo de calificadores?

Su programa deberá recibir de la entrada los parámetros iniciales n y x e imprimir el número con el resultado de ambas preguntas en la salida.


````python
#Desarrolle su código en este espacio

#N --> Número de estudiantes total
N =int( input('Número de estudiantes total: '))
#X --> Número de estudiantes por grupo
X = int(input('Número de estudiantes por grupo: '))
#Ingresar n --> Número de estudiantes; x --> Número de grupos



#Calcular el número de grupos que se pueden formar

N_Grupos = N//X


#Calcular el número de estudiantes sobrantes
N_Grupos_S= N%X

#Calcular el número de reportes
Total_R = N_Grupos*N_Grupos_S

#Imprimir - Número de grupos; numero de reportes
print(f"Hay {N_Grupos} grupos y {Total_R} reportes. ")
````

Podemos calcular este valor a partir de algunos parámetros iniciales que describen el problema. Estos parámetros iniciales son:


La velocidad inicial v0 del objeto en dirección del recorrido.
La posición inicial x0. En este caso el recorrido solo se dará en una dirección de forma horizontal.
La aceleración a que se mantiene constante en todo el recorrido.
El tiempo t o duración del recorrido.

Se conoce que la posición de un objeto uniformemente acelerado puede ser calculada por la siguiente ecuación:


x=x0+v0t+1/2at^2


En este ejercicio usted deberá crear un programa que realice lo siguiente:


1. Obtener de la entrada del programa los parámetros iniciales, línea por línea y en el orden descrito en la sección ⌨️ Entrada.

3. Utilizar los valores numéricos en una expresión matemática y obtener el valor de la posición final.

Donde {x} corresponde a la posición calculada en metros, que deberá ser presentada con un formato que muestre exactamente dos dígitos decimales.

````python
#Desarrolle su código en este espacio

print("""¡Bienvenido al programa!

Ingrese por favor los siguientes datos solicitados
para hallar la posición final del vehículo.

""")
# Se capturan los valores inciales para ingresar en la expresión
X_0 = float(input("Ingrese la posición inicial (m): "))
V_0 = float(input("Ingrese la velocidad inicial (m/s): "))
t = float(input("Ingrese el tiempo final (s): "))
a = float(input("Ingrese la aceleración (m/s): "))


# Se convierten los valores a decimales
#print('{:.2f}'.format(X_0))

#Calcular la expresión numérica para obtener la posición final

X_f = (X_0 + (V_0 * t) + ((1/2)* a * t**2))

#Mostar el resultado final
print('El resultado de la posición final es: {:.2f}'.format(X_f))
````

En este cuarto ejercicio, vamos a adicionar unidades a los valores y variables expresados en la parte anterior del problema
En este caso, el programa va a obtener los parámetros en las siguientes unidades:
Posición inicial x0 => metros (m)
Velocidad inicial v0 => kilómetros x hora (km/h)
La aceleración a => metros sobre segundos al cuadrado (m/s^2)
Tiempo t => segundos (s)

Calcular la posición y la velocidad final.


x=x0+v0t+1/2*(at^2) \
v=v0+a*t^2

Su programa deberá recibir de la entrada 4 lineas con los valores x0,v0,a y t respectivamente.

**Línea 1**:cadena de texto con la posición inicial x0 en m\
**Linea 2**:cadena de texto con la velocidad inicial v0 en km/h\
**Línea 3**:cadena de texto con la aceleración a en m/s^2\
**Línea 4**:cadena de texto con el tiempo t en s

Considere que cada una de estas líneas obtenidas como cadenas de texto con un formato que debe ser convertido a un valor decimal con la función float. No olvide realizar la conversión del tipo de dato de su programa.



````python
#Desarrolle su código en este espacio


#Obtener la entrada del programa los parámetros iniciales
x0 =float( input('Posicion Inicial(m): '))
v0 = float( input('Velocidad Inicial(Km/h): '))
a = float( input('Aceleracion(m/s^2): '))
t = float( input('Tiempo(s): '))
#Realizar las operaciones matemáticas para la conversión de unidades necesarias
v0_1= float((v0*1000)/3600)
#Calculo de las variables
x=(x0+(v0_1*t)+(1/2*(a*(t**2))))
v=(v0_1+(a*(t**2)))

#Imprimir valores
print(f"Posicion Final: {x} m y velocidad final: {v} km/h. ")
````
